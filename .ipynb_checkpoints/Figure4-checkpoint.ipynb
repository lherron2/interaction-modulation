```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
import os
```

```python
plt.rcParams.update({'font.size': 8})
plt.rc('font', family='Helvetica')
```

```python
def param_from_fname(fname, param):
    s = [x.split("=") for x in fname.split("/")[-1].split("_")]
    for idx, value in enumerate(s):
        if value[0] == param:
            return value[1]
        
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

def aspace(start, stop, spacing, decimals=3):
    space = np.linspace(start, stop, int(round((stop - start)/spacing ,2)+1))
    rounded_space = np.around(space, decimals=decimals)
    return rounded_space
```

```python
phase_range_dict={
                "SK": [(-0.5, 1), (0,3)],
                "MAI": [(-0.5, 1), (0,3)],
                "MSI": [(-0.3, 0.3), (0,0.5)],
                "HU": [(-0.5, 1.5), (0,2.0)]
                 }

major_tick_labels={
                "SK": ([0, 1], [0,1,2,3]),
                "MAI": ([0, 1],[0,1,2,3]),
                "MSI": ([-0.2,0,0.2],[0,0.5]),
                "HU": ([0,1],[0,1,2])
                 }

minor_tick_labels={
                "SK": ([-0.5, 0.5], [0.5, 1.5, 2.5]),
                "MAI": ([-0.5, 0.5],[0.5, 1.5, 2.5]),
                "MSI": ([],[0.25]),
                "HU": ([-0.5, 0.5],[0.5,1.5])
                 }
```

## Loading and plotting phase diagrams

```python
fig, axs = plt.subplots(4,6, figsize=(6.5,5.2), dpi=400, tight_layout=True)
net_types = ["SK", "HU", "MAI", "MSI"]
data_path = f"phase_data/"
# plt.axis('off')
for j, net in enumerate(net_types):
    fname = f'nettype={net}_phase_space.npy'
    phase_space_dict = np.load(os.path.join(data_path, fname), allow_pickle=True).item()
    fname = os.path.splitext(fname)[0]
    (theta_min, theta_max) = phase_space_dict["theta"]
    (gamma_min, gamma_max) = phase_space_dict["gamma"]
    (a_min, a_max) = phase_space_dict["a"]
    gt_spacing = phase_space_dict["gt_spacing"]
    a_spacing = phase_space_dict["a_spacing"]
    phase_space = phase_space_dict["phase_space"]

    [(t_min, t_max), (g_min, g_max)] = phase_range_dict[net]
    t_range = aspace(t_min, t_max, gt_spacing)
    theta_range = aspace(theta_min, theta_max, gt_spacing)
    g_range = aspace(g_min, g_max, gt_spacing)
    gamma_range = aspace(gamma_min, gamma_max, gt_spacing)
    (_, _, g_indices) = np.intersect1d(g_range, gamma_range, return_indices=True)
    (_, _, t_indices) = np.intersect1d(t_range, theta_range, return_indices=True)
    phase_space = phase_space[min(t_indices):max(t_indices),min(g_indices):max(g_indices),:]
    
    a_arr = aspace(a_min, a_max, a_spacing)
    ssa = a_arr[:].copy()

    for i, a in enumerate(ssa):
        ax = axs[j][i]
        ax.set_box_aspect(1)
        idx = find_nearest(a_arr,a)
        extent = [t_min, t_max, g_min, g_max]
        im = ax.imshow(phase_space[:,:,idx].T, aspect="auto", 
                         extent=extent, 
                         alpha=1, vmin=0, cmap='Greys', origin="lower", interpolation='None')
        ps_a = phase_space[:,:,idx].T.copy()
        ps = ps_a > 0.9
        ps_a[ps] = 1
        ps_a[~ps] = 0
        
        C = ax.contour(ps_a, levels=0, alpha=1, 
                      origin='lower', cmap='flag', 
                      extend='both', linewidths=1.25, extent=extent,
                    zorder=1, antialiased=True)
    
        axs[j][-1].imshow(ps_a, aspect="auto", 
                                 extent=extent, 
                                 alpha=0.25, vmin=0, cmap='Greys', origin="lower", interpolation="None")
        axs[j][-1].set_box_aspect(1)
        axs[j][i].set_xticks(ticks=major_tick_labels[net][0])
        axs[j][i].set_xticks(ticks=minor_tick_labels[net][0], minor=True)
        axs[j][i].set_yticks(ticks=major_tick_labels[net][1])
        axs[j][i].set_yticks(ticks=minor_tick_labels[net][1], minor=True)
        axs[j][i].set_yticklabels([])
        axs[j][i].set_xticklabels([])

        if i == 0:
            axs[j][i].set_ylabel(r"$\lambda$", size=8)
            axs[j][i].set_yticks(ticks=major_tick_labels[net][1])
            axs[j][i].set_yticklabels(major_tick_labels[net][1])

        else:
            axs[j][i].set_yticklabels([])
        
        if j == 3:
            axs[j][i].set_xlabel(r"$\theta$", size=8)
        
        axs[j][-1].set_yticks(ticks=minor_tick_labels[net][1], minor=True)
        axs[j][-1].set_yticks(ticks=major_tick_labels[net][1])

        axs[j][-1].set_yticklabels([])
        axs[j][-1].set_xticks(ticks=minor_tick_labels[net][0], minor=True)
        axs[j][-1].set_xticks(ticks=major_tick_labels[net][0])
        axs[j][-1].set_xticklabels(major_tick_labels[net][0])
        
        axs[3][-1].set_xlabel(r"$\theta$", size=8)

for ax, col in zip(axs[0,:], ssa):
    ax.annotate(f"$a = {col}$", (0.5, 0.9), xytext=(0, 10), ha='center', va='bottom',
                 xycoords='axes fraction', textcoords='offset points', fontsize=8)
    
axs[0][2].scatter([0.35], [1.25], color="lightgreen", alpha=1, s=10, zorder=3)
axs[1][2].scatter([0.62], [0.3], color="lightgreen", alpha=1, s=10, zorder=3)
axs[2][2].scatter([0.3], [1.5], color="lightgreen", alpha=1, s=10, zorder=3)
axs[3][2].scatter([0.05], [0.1], color="lightgreen", alpha=1, s=10, zorder=3)

axs[0,-1].annotate(f"Intersection", (0.5, 0.9), xytext=(0, 10), ha='center', va='bottom',
                 xycoords='axes fraction', textcoords='offset points', fontsize=8)
fig.align_ylabels(axs[:, 0])
plt.subplots_adjust(wspace=0.3,hspace=-0.57)
```
